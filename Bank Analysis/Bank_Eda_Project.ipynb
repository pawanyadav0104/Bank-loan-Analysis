```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\pawan\OneDrive\Desktop\Banking.csv')
df.head(5)              
```

```python
# genrating descriptive statics for the dataframe
df.describe()
```

```python
df['Estimated Income'].min()
```

```python
bins=[0,100000,300000,float('inf')]
labels=['Low','Med','high']

df['Income Band']= pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Examine the distribution in unqiue categories
categorical_cols=df[["BRId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]

for col in categorical_cols:
    print(f"Value Counts for'{col}':")
    display(df[col].value_counts())
```

```python
for i, predictor in enumerate(df[["BRId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='GenderId')
```

```python
for i, predictor in enumerate(df[["BRId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='Nationality')
```

```python
# histogram for different Occupation
for col in categorical_cols:
    if col=="Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
#numerical analysis
numerical_cols=['Estimated Income','Superannuation Savings']
plt.figure(figsize=(10,10))
for i,col in enumerate(numerical_cols):
    plt.subplot(2,1,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

```python
#heatmaps
```

```python
correlation_matrix=df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

Insights of EDA

The strongest Positive correlation occur among "Bank Deposits" with "checking Acc" , "Saving acc" indicating that customer who maintain high balances in one account types often hold substanial amount across other accounts as well
